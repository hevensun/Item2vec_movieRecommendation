```python
import pandas as pd
import numpy as np

#df_movies = pd.read_csv('ml-20m/movies.csv')
df_ratings = pd.read_csv('ml-20m/ratings.csv')
```

```python
def rating_splitter(df):
    
    df['liked'] = np.where(df['rating']>=5, 1, 0)
    df = df[df.liked == 1]
    df['movieId'] = df['movieId'].astype('str')
    gp_user_like = df.groupby(['liked', 'userId'])
    return gp_user_like
def getMovies(gp_user_like):
    return ([gp_user_like.get_group(gp)['movieId'].tolist() for gp in gp_user_like.groups])
pd.options.mode.chained_assignment = None
gp_user_like = rating_splitter(df_ratings)
splitted_movies = getMovies(gp_user_like)
print(len(splitted_movies))
```

```python
df_movies = pd.read_csv('ml-20m/movies.csv')


movieId_to_name = pd.Series(df_movies.title.values, index = df_movies.movieId.values).to_dict()
name_to_movieId = pd.Series(df_movies.movieId.values, index = df_movies.title).to_dict()
```

```python
MovieSet = [[movieId_to_name[int(id)] for id in basket]for basket in splitted_movies]
```

```python
print(MovieSet[:2])
```

```python
from gensim import corpora, models
dictionary = corpora.Dictionary(MovieSet)
# print(dictionary)
# Dictionary(32 unique tokens: [u'often', u'feel', u'profession', u'drive', u'say']...)

# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in MovieSet]
```

```python
import gensim
Ntopic = 3
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=Ntopic, id2word=dictionary, passes=20)
print(ldamodel.print_topics(num_topics=Ntopic, num_words=10))
```

```python
userLikeSciFi = []
for user in usersLikeMovie:
    bow = dictionary.doc2bow(user)
    topic_probability = ldamodel.get_document_topics(bow)
    topicIndex = 3
    sciProb = [pair for pair in topic_probability if pair[0]==topicIndex]
    #print(sciProb)
    if sciProb ==[]:continue
    #print(sciProb[0][1])
    if sciProb[0][1]>0.8:
        userLikeSciFi.append(user)
```
